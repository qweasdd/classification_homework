```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('max_rows', 10)
pd.options.display.float_format = '{:.1f}'.format
```

```python
data = pd.read_csv('vibory_data.csv')
```

```python
data
```

```python
data['Явка'] = data.iloc[:, 12] / data.iloc[:, 3] * 100
```

```python
data.sort_values('Явка', ascending= False)
```

выберите произвольного кандидата и найдите тот избиратльный участок, на котором он получил наибольший результат (%, учитывать участки на которых проголосовало больше 300 человек)

Сурайкин

```python
temp = data[data.iloc[:, 12] > 300]
```

```python
(temp.iloc[:, 20] / temp.iloc[:, 12] * 100).max()
```

```python
idxmax = (temp.iloc[:, 20] / temp.iloc[:, 12] * 100).idxmax()
```

```python
data.iloc[idxmax, [0, 1, 2, 20]]
```

найдите регион, где разница между ТИК с наибольшей явкой и наименьшей максимальна

```python
temp = data.groupby(['Регион', 'ТИК']).apply(lambda x : x.iloc[:, 12].sum()).to_frame().reset_index() \
.groupby('Регион')[0].apply(lambda x : x.max() - x.min())
```

```python
temp.max()
```

```python
temp.idxmax()
```

посчитайте дисперсию по явке для каждого региона (учитывать УИК)

```python
data.groupby('Регион')['Явка'].var()
```
